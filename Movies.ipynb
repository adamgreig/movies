# Movie RecommendationI keep track of my movie collection using [Griffith](http://www.griffith.cc/), which can output a CSV of my movie data.Specifically it tracks which films I have or have not seen, and what rating I gave the films. I use the following rating scale, derived from an initial score of 0-5:<table>    <tr><td>0</td><td>Awful</td></tr>    <tr><td>2</td><td>Didn't Like</td></tr>    <tr><td>4</td><td>OK</td></tr>    <tr><td>6</td><td>Liked</td></tr>    <tr><td>8</td><td>Really Liked</td></tr>    <tr><td>10</td><td>Amazing</td></tr></table>Typically I'll have a number of as-yet unseen films in my collection, and deciding which film to watch next is a challenge!Clearly the best solution is to make a system to choose for me.We'll use [NumPy](http://numpy.org) (of course) and [scikit-learn](http://scikit-learn.org) (provides the machine learning),alongside [pandas](http://pandas.pydata.org) since I haven't used it before and the best way to learn is by doing!

Load the films CSV from Griffith. Each entry includes the title, whether the user has seen the film, the user rating (0-10) and a pipe-and-whitespace separated list of genres.

```python
films = pd.read_csv("https://randomskk.net/u/movie_collection/films.csv", index_col="title").sort()
```

## Parsing GenresFirst we'll obtain a list of all the genres in use. For each genre list, split and strip the genres, reduce the resulting lists and take the unique set.

```python
genres = list(set(films["genre"].map(lambda genre: [g.strip() for g in genre.split("|")]).sum()))genres.sort()print ", ".join(g for g in genres)
```

For each genre, we'll add a boolean column to the DataFrame to indicate whether that film belongs to that genre.

```python
films[genres] = pd.DataFrame([films["genre"].map(lambda g: genre in g) for genre in genres], index=genres).Tfilms[genres[:10]].head(3)
```

For convenience, store a view of seen and unseen films.

```python
seen_films = films[films["seen"]]unseen_films = films[-films["seen"]]
```

Check out the stats for the ratings per genre. This is assuming that membership of one genre is independent of the rest.

```python
# Replace False with NaN and True with the respective film's rating so we can easily compute per-genre statisticsgenre_stats = (seen_films[genres].replace(False, np.nan) * seen_films["rating"].values.reshape((len(seen_films), 1)).repeat(len(genres), axis=1))plt.figsize(10, 8)genre_stats.hist(range=(0, 10), bins=6, sharex=True, sharey=True)plt.xlim((0, 10))genre_stats.describe().T.sort_index(by="mean", ascending=False)
```

## Machine LearningNow train a Random Forest regressor on the films that have been seen. The boolean vector of genres is the feature vector, while the rating is the variable to be regressed.Then predict the rating for all the unseen films.

```python
training = films[films["seen"]][genres + ["rating"]]classifier = RandomForestRegressor(n_estimators=1000, n_jobs=-1)classifier.fit(training[genres], training["rating"])unseen_films = films[-films["seen"]]unseen_films["predicted_rating"] = classifier.predict(unseen_films[genres])unseen_films.sort_index(by="predicted_rating", ascending=False)[["predicted_rating", "genre"]].head(10)
```
